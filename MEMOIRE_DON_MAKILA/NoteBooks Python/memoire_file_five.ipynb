```python
# Importation des bibliothèques
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_test_No_NaN.csv')
```

```python
df.head()
```

```python
# Séparer le dataset en colonnes numériques et catégorielles
df_nums = df.select_dtypes(exclude='object')
df_objs = df.select_dtypes(include='object')
```

```python
df_nums.info()
```

```python
df_nums.describe().transpose()
```

```python
df_objs.info()
```

```python
df_objs.describe().transpose()
```

```python
# Transformer les types object en numérique
```

```python
# La colonne Gender : {'Male':1,'Female':0}
gender_value={'Male':1,'Female':0}
df_objs['Gender'] = df_objs['Gender'].map(gender_value)
```

```python
# La colonne Married : {'Yes':1,'No':0}
married_values = {'Yes':1,'No':0}
df_objs['Married'] = df_objs['Married'].map(married_values)
```

```python
# La colonne Dependents : {'0':0,'1':1,'2':2,'3+':3}
dependents_values = {'0':0,'1':1,'2':2,'3+':3}
df_objs['Dependents'] = df_objs['Dependents'].map(dependents_values)
```

```python
# La colonne Education : {'Graduate':1,'Not Graduate':0}
education_values = {'Graduate':1,'Not Graduate':0}
df_objs['Education'] = df_objs['Education'].map(education_values)
```

```python
# La colonne Self_Employed : {'Yes':1,'No':0}
self_Employed_values = {'Yes':1,'No':0}
df_objs['Self_Employed'] = df_objs['Self_Employed'].map(self_Employed_values)
```

```python
# La colonne Property_Area : {'Rural':0,'Urban':1,'Semiurban':2}
property_Area_values = {'Rural':0,'Urban':1,'Semiurban':2}
df_objs['Property_Area'] = df_objs['Property_Area'].map(property_Area_values)
```

```python
df_objs
```

```python
df = pd.concat([df_objs,df_nums],axis=1)
```
